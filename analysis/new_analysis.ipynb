```python
import pandas as pd

df = pd.read_csv("ache_donepezil_contacts_9poses.csv")

contact_freq = (
    df.groupby(["resn", "resi", "chain"])
      .size()
      .reset_index(name="contact_count")
      .sort_values("contact_count", ascending=False)
)

contact_freq.head(15)
```

```python
import matplotlib.pyplot as plt

top15 = contact_freq.head(15)

plt.figure(figsize=(8,5))
plt.bar(
    [f"{r.resn}{r.resi}" for _, r in top15.iterrows()],
    top15["contact_count"]
)
plt.xlabel("Residue")
plt.ylabel("Contact frequency (poses)")
plt.title("Top 15 AChE residues contacting Donepezil (9 poses)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
